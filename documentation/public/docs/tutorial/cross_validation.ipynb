## What is cross-validation?


[Cross-validation](https://en.wikipedia.org/wiki/Cross-validation_(statistics)) is a method for estimating the quality of a model. It serves the same purpose as the **train and test protocol**, but cross-validation is more accurate.

Cross-validation metrics can be estimated with the `cross_validation` method. Here is an example of how to evaluate a learner using cross-validation:

**Note:** Cross-validation only requires a single dataset. If you have a training and testing dataset sampled from the same distribution, it is best to merge both datasets and to apply cross-validation on the result.

```python
import ydf
import pandas as pd

# Download a classification dataset and load it as a Pandas DataFrame.
ds_path = "https://raw.githubusercontent.com/google/yggdrasil-decision-forests/main/yggdrasil_decision_forests/test_data/dataset"
dataset = pd.read_csv(f"{ds_path}/adult.csv")

# Print the first 5 examples
dataset.head(5)
```

```python
learner = ydf.RandomForestLearner(label="income")
evaluation = learner.cross_validation(dataset, folds=10)

evaluation
```
